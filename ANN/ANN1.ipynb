```python


class ThresholdHidden(Layer):

    def __init__(self, threshold_values, **kwargs):
        super(ThresholdHidden, self).__init__(**kwargs)
        self.threshold_values = threshold_values

    def call(self, inputs):
        return K.cast(K.greater(inputs, self.threshold_values), K.floatx())

    def get_config(self):
        config = {'threshold_values': self.threshold_values}
        base_config = super(ThresholdHidden, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))
```
